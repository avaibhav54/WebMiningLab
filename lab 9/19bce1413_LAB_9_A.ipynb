```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph([("A", "B"), 
                ("A", "F"), 
                ("C", "B"),
                ("D", "A"), 
                ("D", "E"), 
                ("D", "G"), 
                ("E", "B"), 
                ("E", "C"), 
                ("E", "H"), 
                ("F", "E"), 
                ("G", "A"), 
                ("G", "B"), 
                ("G", "C"), 
                ("G", "F"), 
                ("H", "B"), 
                ("H", "D")])
nx.draw_circular(G, with_labels=True, node_color="#008000", node_size=800)
```

```python
P_rank = nx.pagerank(G)
P_rank
```
